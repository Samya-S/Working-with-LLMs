```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.together import TogetherLLM
import os
import re

# Set your Together AI API key
os.environ["TOGETHER_API_KEY"] = TOGETHER_API_KEY

# Load the TXT file(s)
documents = SimpleDirectoryReader("D:/GitHub/Working with LLMs/Session 1/data").load_data()

# Set global settings (Replaces ServiceContext)
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en")
Settings.llm = TogetherLLM(
    model=LLM_MODEL, 
    temperature=0.3,
)

# Create an index
index = VectorStoreIndex.from_documents(documents)

# Create a query engine
query_engine = index.as_query_engine()

# Ask a question about the text
query = "Give the 3 emails of the people in the document"
response = query_engine.query(query)

# Remove the <think>...</think> portion
cleaned_text = re.sub(r"<think>.*?</think>", "", str(response), flags=re.DOTALL).strip()

# Print cleaned output
print(cleaned_text)
```
